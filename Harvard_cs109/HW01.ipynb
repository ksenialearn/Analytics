```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,30) #30 points from 0 to 10
y = np.sin(x)
plt.plot(x, y, 'ro-', label = 'A sine wave')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc = 'lower left')
```
