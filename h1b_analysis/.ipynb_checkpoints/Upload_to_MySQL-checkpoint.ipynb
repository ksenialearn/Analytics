```python
import mysql.connector
from mysql.connector import errorcode
import pandas as pd
```

```python
data = pd.read_csv('h1b.csv')
data.head()
```

```python
data.dtypes
```

```python
def get_conn():
    return mysql.connector.connect(host= "localhost",
                  user="root",
                  passwd="password",
                  db="h1bdata")
```

```python
TABLES = {}
TABLES['data'] = (
    "CREATE TABLE `data` ("
    "  `id` int NOT NULL,"
    "  `CASE_STATUS` varchar(200),"
    "  `EMPLOYER_NAME` varchar(200),"
    "  `SOC_NAME` varchar(200),"
    "  `JOB_TITLE` varchar(200),"
    "  `FULL_TIME_POSITION` varchar(200),"
    "  `PREVAILING_WAGE` float(12,2),"
    "  `YEAR` int,"
    "  `WORKSITE` varchar(200),"
    "  `lon` float(12,6),"
    "  `lat` float(12,6),"
    "  PRIMARY KEY (`id`), UNIQUE KEY `id` (`id`)"
    ") ENGINE=InnoDB")

conn = get_conn()
cursor = conn.cursor()
for name, ddl in TABLES.iteritems():
    try:
        print("Creating table {}: ", name)
        cursor.execute(ddl)
    except mysql.connector.Error as err:
        if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
            print("already exists.")
        else:
            print(err.msg)
    else:
        print("OK")
conn.commit()
cursor.close()
conn.close()
```

```python
headers = 'CASE_STATUS, EMPLOYER_NAME, SOC_NAME, JOB_TITLE, FULL_TIME_POSITION, PREVAILING_WAGE, YEAR, WORKSITE, lon, lat'
headers = headers.split(", ")
print(headers)
```

```python

def create_insert_sql(line, headers, id):
    """
    Create a SQL statement for inserting row 
    """
    values = []
    for header in headers:
        val = line[header]
        if header == 'YEAR': val = int(val)
        if repr(val) == 'nan':
            values.append('null')
        else:
            values.append(repr(val))

    headerpart = ', '.join(headers)    
    valuespart = ', '.join(values)

    sql = 'INSERT INTO h1bdata.data (' + 'id, ' + headerpart + ')' + ' VALUES' + '(' + str(id) + ', ' + valuespart + ')'
    return sql



line = data.iloc[2,:]
print(create_insert_sql(line, headers, 123))
```

```python
#[leng, wid] = data.shape

conn = get_conn()
cursor = conn.cursor()

try:
    for row in range(0,data.shape[0]):
        if row % 1000 == 0: print(row)
        line = data.iloc[row,:]
        sql = create_insert_sql(line, headers, row+1)
        try:
            cursor.execute(sql)
        except Exception as e:
            print(sql)
            raise e
    print("all done")
    conn.commit()
    
finally:
    cursor.close()
    conn.close()
```
